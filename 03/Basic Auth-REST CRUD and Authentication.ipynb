```python
import requests
import pandas as pd
from io import StringIO
from html.parser import HTMLParser

class MLStripper(HTMLParser):
    def __init__(self):
        super().__init__()
        self.reset()
        self.strict = False
        self.convert_charrefs= True
        self.text = StringIO()
    def handle_data(self, d):
        self.text.write(d)
    def get_data(self):
        return self.text.getvalue()
    
def strip_tags(html):
    s = MLStripper()
    s.feed(html)
    return s.get_data()
    

user = 'admin'
url = 'http://localhost/wp_hairdye/wp-json/wp/v2/hair_dyes'
r = requests.get(url)
print("Status Code:",r.status_code)
r_list = r.json()

dataset = []
dict = {}
for i in r_list:    
     # print(i['id'])
     dict['id'] = i['id']   
     title = i['title']
     content = i['content']
     for k, v in title.items():
        # print(v)
        dict['title'] = v
     for k, v in content.items():
        # print(v)
        if k == 'rendered':
            dict['content'] = strip_tags(v)
     dataset.append(dict.copy())
print(dataset)
print('**************************************************************************')
print('\n')
print(pd.DataFrame(dataset))
```
